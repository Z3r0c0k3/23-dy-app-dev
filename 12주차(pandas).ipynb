## 데이터 병합

```python
import pandas as pd

# 첫 번째 DataFrame 생성
data1 = pd.DataFrame({
 'Name': ['Kim', 'Lee', 'Park'],
 'Age': [25, 30, 22]
})

# 두 번째 DataFrame 생성
data2 = pd.DataFrame({
 'Name': ['Kim', 'Lee', 'David'],
 'Gender': ['Female', 'Male', 'Male']
})

# 'Name' 열을 기준으로 두 DataFrame을 병합
merged_data = pd.merge(data1, data2, on='Name', how='outer')
print(merged_data)


# how='inner' : 두 데이터프레임 모두에 존재하는 것만
# outer : 두 데이터프레임 데이터 전체
```

## Lab:   판다스 데이터 처리 (p.539)

```python
# 상품 판매 데이터를 다루는 종합 예제를 만들어보자.
# 이 예제에서는 판다스의 데이터 프레임을 생성하고,
# 데이터 정렬, 필터링, 그룹화, 데이터 변형, 병합 등
# 다양한 데이터 처리와 분석 작업을 수행한다.

import pandas as pd

# 물건 판매 데이터 생성
data = pd.DataFrame({
   'Date': ['2023-07-01', '2023-07-01', '2023-07-02', '2023-07-02', '2023-07-03'],
   'Product': ['A', 'B', 'A', 'C', 'B'],
   'Price': [1000, 1500, 800, 2000, 1200],
   'Quantity': [5, 3, 7, 2, 4]
})

# 데이터 출력
print("=== 원본 데이터 ===")
print(data)

# 1. 데이터 정렬
sorted_data = data.sort_values(by='Price', ascending=False)
print("\n=== 데이터 정렬 ===")
print(sorted_data)

# 2. 데이터 필터링
filtered_data = data[data['Price'] >=1000]
print("\n=== 데이터 필터링 ===")
print(filtered_data)

# 3. 데이터 그룹화와 집계
grouped_data = data.groupby('Product')['Price'].mean()
print("\n=== 데이터 그룹화와 집계 ===")
print(grouped_data)

# 4. 데이터 변형
data['TotalPrice'] = data['Price'] *data['Quantity']
print("\n=== 데이터 변형 ===")
print(data)

# 5. 데이터 병합
additional_data = pd.DataFrame({
   'Date': ['2023-07-03', '2023-07-04'],
   'Product': ['A', 'B'],
   'Price': [900, 1300],
   'Quantity': [6, 5]
})
merged_data = pd.concat([data, additional_data], ignore_index=True)
print("\n=== 데이터 병합 ===")
print(merged_data)
```

## statsmodels라이브러리의 OLS 이용한 단순회귀분석

```python
import statsmodels.formula.api as smf
```

## 1. 모델 구축

```python
# 종란무게 - 병아리무게 단순선형회귀모델 구축
model_lm = smf.ols(formula = 'weight ~ egg_weight', data = w2)
```

## 2. 모델 학습

```python
result_lm = model_lm.fit()
```

## 3. 모델 결과 확인

```python
result_lm.summary()
```
